### **Business Problem Understanding**

**Context**

Pelanggan selalu menjadi teka-teki bagi sebuah bisnis. Banyak sekali buku-buku mengenai cara sukses sebuah bisnis, jika *deep dive* kedalamnya semua akan kembali lagi ke pelanggan, karena merekalah yang membawa masuk keuntungan. Akan tetapi tidak semua pelanggan sesuai dengan bisnis, beberapa pelanggan mungkin membawa nilai lebih bagi bisnis dan beberapa mungkin tidak.

Customer Lifetime Value, atau CLV, adalah ukuran seberapa berharganya seorang pelanggan bagi sebuah perusahaan. CLV hadir sebagai *tools* yang membantu untuk mengidentifikasi pelanggan yang *valuable* bagi bisnis. CLV dapat dimanfaatkan secara efektif dengan mengidentifikasi pelanggan yang bernilai tinggi dengan memahami perilaku, preferensi, dan kebutuhan mereka. 

[*"CLV" The Only Metric That Matters* - "Forbes"](https://www.forbes.com/sites/homaycotte/2015/08/25/customer-lifetime-value-the-only-metric-that-matters/)

Dengan menggunakan CLV (Customer Lifetime Value), perusahaan memiliki dasar untuk merumuskan dan menerapkan strategi yang spesifik untuk pelanggan, guna memaksimalkan keuntungan "*lifetime*" dari setiap pelanggan serta bisa memperpanjang durasi hubungan dengan pelanggan tersebut.

**Problem Statement**

Perusahaan *AUTOLIFE Insurance* sebagai salah satu perusahaan asuransi automotif asal US, memiliki kesulitan dalam mengidentifikasi seluruh pelanggan. **Dengan resource yang ada, perusahaan kesulitan dalam mengidentifikasi pelanggan mana yang menjanjikan potensi tertinggi untuk profitabilitas jangka panjang. Hal yang menjadi tantangan utama adalah cara memprediksi CLV pelanggan secara akurat**. Ini akan memungkinkan perusahaan untuk lebih efektif menargetkan pelanggan yang paling berharga dan merencanakan pertumbuhan jangka panjang dengan lebih baik.

Tim marketing dari *AUTOLIFE Insurance* berperan melakukan kampanye pemasaran yang menargetkan pelanggan yang ada, seperti promosi, komunikasi rutin, dan penawaran khusus. Maka dari itu, mereka sangat membutuhkan cara bagaimana dengan lebih efektif menentukan pelanggan yang berharga ini.

**Goals**|

Berdasarkan permasalahan yang ada, perusahaan *AUTOLIFE Insurance* tentunya ingin memiliki sebuah *tools* yang dapat memilih pelanggan mana yang *valuable* bagi bisnis mereka. Bagi tim marketing, adanya *tools* tersebut akan sangat membantu dalam menentukan dasaran yang dapat diukur dalam menentukan pelanggan-pelanggan yang memiliki potensi keuntungan jangka panjang, dan berkelanjutan. 

Harapannya tim marketing dapat **memperkirakan nilai CLV untuk setiap pelanggan secara akurat** sehingga membantu dalam membuat keputusan strategi marketing untuk memaksimalisasi keuntungan dalam jangka panjang. Serta dapat **mengidentifikasi faktor-faktor utama yang mempengaruhi CLV** sehingga dapat mengambil keputusan berbasis data dalam mengelola hubungan dengan pelanggan.

**Analytic Approach**

Kita akan menyiapkan dasaran untuk menganalisis data sehingga dapat menemukan pola dari fitur-fitur yang ada dari data existing klien. Semua itu akan berdasarkan CLV yang sudah terkalkulasi dan beberapa variabel-variabel pendukung.

Evaluasi metrik yang akan digunakan adalah RMSE, MAE, dan MAPE, di mana RMSE adalah nilai rataan akar kuadrat dari error, MAE adalah rataan nilai absolut dari error, sedangkan MAPE adalah rataan persentase error yang dihasilkan oleh model regresi. Semakin kecil nilai RMSE, MAE, dan MAPE yang dihasilkan, berarti model semakin akurat dalam memprediksi nilai Customer Lifetime Value sesuai dengan limitasi fitur yang digunakan.

**Outliers Handling**

```python
# Regular functionnya
def Outliers_Check(df, cols):
    outlier_summary = []

    for col in cols:
        # Menggunakan quantile() untuk Q1 dan Q3
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        Upper_Fence = Q3 + (1.5 * IQR)
        Lower_Fence = Q1 - (1.5 * IQR)

        df_outliers = df[(df[col] < Lower_Fence) | (df[col] > Upper_Fence)]
        num_outliers = len(df_outliers)

        outlier_summary.append({
            'Column': col,
            'Lower Fence': Lower_Fence,
            'Upper Fence': Upper_Fence,
            'Num of Outliers': num_outliers
        })

        # Membuat boxplot untuk kolom saat ini
        plt.figure(figsize=(6, 2))  
        sns.boxplot(x=df[col])  
        plt.title(f'Box Plot for {col}')  
        plt.xlabel(col)  
        plt.tight_layout() 
        plt.show()

    return pd.DataFrame(outlier_summary)
```

> Sekarang coba kita cek outliers pada data numerikal (`Number of Policies`, `Monthly Premium Auto`, `Total Claim Amount`, `Income`, `Customer Lifetime Value`):
>
> Pertama, mari kita cek `Number of Policies`

**-> Outlier `Number of Policies`**

```python
outlier_results = Outliers_Check(df, ['Number of Policies'])
outlier_results
```

* `Number of Policies` kolom menunjukan jumlah polis asuransi dari pelanggan yang berkisar antara 1-9, perlu diketahui jumlah polis tidak bermasalah memiliki outlier selama masih masuk akal dengan domain knowledge. Dapat diketahui bahwa policy pada asuransi mobil cukup banyak di US [source: [[1]](https://www.iii.org/fact-statistic/facts-statistics-auto-insurance),[[2]](https://content.naic.org/consumer/auto-insurance.htm)] . Maka, tidak diperlukan handling pada kolom ini.

**Data Distribution Check**

Analisa distribusi data dapat melihat bagaimana kondisi distribusi dari data, apakah normal atau tidak normal. Dengan melakukan uji normalitas, kita dapat memilih metode analisis yang tepat dan melakukan transformasi data jika diperlukan, sehingga memastikan hasil yang lebih akurat dan dapat diandalkan.

```python
# Function to perform normality test
def normality_test(column):
    stat, p_value = stats.shapiro(column)
    return stat, p_value

# Applying the normality test to each column
results = {}
for column in ['Number of Policies','Monthly Premium Auto','Total Claim Amount','Income','Customer Lifetime Value']:
    stat, p_value = normality_test(df[column])
    normality_result = 'Normal' if p_value > 0.05 else 'Not Normal'  # Determine normality
    results[column] = {
        'Distribution': normality_result
    }

# Display results
results_df = pd.DataFrame(results).T
results_df
```

**Import Libraries Modeling**

```python
# Preprocessing Modeling
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
import category_encoders as ce
from sklearn.preprocessing import RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Machine learning
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost.sklearn import XGBRegressor

# Evaluation
from sklearn.model_selection import cross_validate
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
```

**Data Encoding**

Encoding adalah salah satu tahap praproses data sebelum diproses dengan algoritma machine learning. Maka, akan sangat mungkin menemukan satu atau beberapa fitur yang bertipe kategori.Untuk keperluan pengolahan data, data-data tersebut harus dinyatakan ke dalam format numerik (encoding) supaya dapat diolah menggunakan formula matematika.

Untuk project ini, kita akan menggunakan One-Hot Encoding untuk `Vehicle Class`, `EmploymentStatus`, `Renew Offer Type`, `Marital Status` dan juga menggunakan Ordinal Encoding untuk `Education` dan `Coverage`.

```python
# Ordinal map
ordinal_map = [
    {'col': 'Coverage', 'mapping': {'Basic': 1, 'Extended': 2, 'Premium': 3, None:0}},
    {'col': 'Education', 'mapping': {'High School or Below': 1, 'College': 2, 'Bachelor': 3, 'Master': 4, 'Doctor': 5, None:0}}
]
# Menggabungkan encoder dalam column transformer
col_transform = ColumnTransformer([
    ('onehot', OneHotEncoder(drop='first'), ['Vehicle Class', 'EmploymentStatus', 'Renew Offer Type', 'Marital Status']),
    ('ordinal', ce.OrdinalEncoder(mapping=ordinal_map), ['Coverage', 'Education'])  
], remainder='passthrough')

col_transform
```

```python
# Define X dan y
X = df.drop(columns='Customer Lifetime Value')
y = df['Customer Lifetime Value']
```

**Train & Test Splitting**

```python
# Membagi data menjadi set pelatihan dan pengujian 70:30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

**Benchmark Model**

```python
# Define scaler, karena distribusi tidak normal maka menggunakan robust scaler yang menggunakan median dan IQR
robust = RobustScaler()
```

Kandidat Model :
* Kita tetap mencoba LinearRegression sebagai model acuan.
* Untuk data yang memiliki hubungan yang tidak linier dan kompleks, kita akan mencoba model KNeighborsRegressor, DecisionTreeRegressor, dan RandomForestRegressor.
* Untuk data kompleks dan membutuhkan akurasi tinggi, kita akan mencoba model GradientBoostingRegressor dan XGBRegressor.

```python
# Definisikan model
lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state= 0)
rf = RandomForestRegressor(random_state= 0)
xgb = XGBRegressor(random_state= 0)
gbr = GradientBoostingRegressor(random_state= 0)

models = [lr, knn, dt, rf, xgb, gbr]

# Prepare untuk menyimpan hasil evaluasi
score_RMSE = []
list_mean_RMSE = []
list_std_RMSE = []
score_MAE = []
list_mean_MAE = []
list_std_MAE = []
score_MAPE = []
list_mean_MAPE = []
list_std_MAPE = []

# Evaluation Metrics yang digunakan
scoring = {'rmse': 'neg_root_mean_squared_error',
           'mae': 'neg_mean_absolute_error',
           'mape': 'neg_mean_absolute_percentage_error'}

for model in models:
    # Membuat pipeline untuk setiap model
    pipe_model = Pipeline([
        ('prep', col_transform),
        ('scaler', robust),  # Gunakan RobustScaler
        ('algo', model)      # Algoritma yang digunakan
    ])

    # Cross-validation untuk setiap model
    cv_score = cross_validate(
        estimator=pipe_model,
        X=X_train,
        y=y_train,
        scoring=scoring,
        cv=5  # Menggunakan 5-fold cross-validation
    )

    # Menyimpan hasil evaluasi untuk setiap metrik
    score_RMSE.append(cv_score['test_rmse'])
    list_mean_RMSE.append(cv_score['test_rmse'].mean())
    list_std_RMSE.append(cv_score['test_rmse'].std())

    score_MAE.append(cv_score['test_mae'])
    list_mean_MAE.append(cv_score['test_mae'].mean())
    list_std_MAE.append(cv_score['test_mae'].std())

    score_MAPE.append(cv_score['test_mape'])
    list_mean_MAPE.append(cv_score['test_mape'].mean())
    list_std_MAPE.append(cv_score['test_mape'].std())

# Menampilkan hasil
df_results = pd.DataFrame({
    'Model': ['Linear Regression', 'KNN', 'Decision Tree', 'Random Forest', 'XGBoost', 'Gradient Boosting'],
    'Mean RMSE': list_mean_RMSE,
    'STD RMSE': list_std_RMSE,
    'Mean MAE': list_mean_MAE,
    'STD MAE': list_std_MAE,
    'Mean MAPE': list_mean_MAPE,
    'STD MAPE': list_std_MAPE
}).round(3)

# Menampilkan hasil evaluasi
df_results
```

Berdasarkan cross validation diatas, 
- Jika melihat nilai RMSE, Gradient Boost memiliki nilai paling rendah. 
- Berdasarkan nilai MAE dan MAPE, RandomForest memiliki nilai paling rendah.
- Berdasarkan cross-val tingkat kestabilan data (STD) XGBoost paling stabil. 
  
Maka selanjutnya, kita akan dilakukan prediksi pada test set dengan 3 benchmark model terbaik, yaitu Gradient Boost, RandomForest dan XGBoost. 

**Predict to Test Set with the Benchmark Model**

Prediksi pada test set dengan menggunakan model Gradient Boost, RandomForest dan XGBoost.

```python
# Model terbaik untuk evaluasi
test_models = [RandomForestRegressor(random_state=0), GradientBoostingRegressor(random_state=0), XGBRegressor(random_state= 0)]

# Menyimpan hasil evaluasi
nilai_rmse = []
nilai_mae = []
nilai_mape = []

# Prediksi pada test set dengan model terpilih
for model in test_models:
    # Membuat pipeline untuk setiap model
    pipe_model = Pipeline([
        ('prep', col_transform),
        ('scaler', robust),  # Gunakan RobustScaler
        ('algo', model)      # Algoritma yang digunakan
    ])

    # Fitting model dengan training data
    pipe_model.fit(X_train, y_train)

    # Prediksi pada test set
    yPred = pipe_model.predict(X_test)

    # Menghitung RMSE, MAE, dan MAPE
    rmse = np.sqrt(mean_squared_error(y_test, yPred))  # Root Mean Squared Error
    mae = mean_absolute_error(y_test, yPred)           # Mean Absolute Error
    mape = mean_absolute_percentage_error(y_test, yPred)  # Mean Absolute Percentage Error

    # Menyimpan hasil per model
    nilai_rmse.append(rmse)
    nilai_mae.append(mae)
    nilai_mape.append(mape)

# Menampilkan hasil
df_results = pd.DataFrame({
    'Model': ['Random Forest', 'Gradient Boosting', 'XGBoost'],
    'RMSE Base': nilai_rmse,
    'MAE Base': nilai_mae,
    'MAPE Base': nilai_mape
}).round(3)

df_results
```

Berdasarkan model :
* Hasil RMSE pada Gradient Boosting memberikan hasil yang lebih rendah daripada Random Forest dan XGBoost (831.848 vs 859.697 vs 919.561), hal ini menunjukkan model Gradient Boosting lebih baik dalam meminimalisir error yang besar, berarti lebih akurat dalam prediksi.
* berbanding terbalik dengan RMSE, Hasil dari MAE menunjukkan Random Forest lebih unggul dari Gradient Boosting dan XGBoost (343.070 vs 365.040 vs 438.993). Hal ini menunjukkan bahwa Random Forest memiliki absolut error yang lebih kecil. 
* Random Forest memiliki MAPE yang sedikit lebih rendah dibandingkan Gradient Boosting dan XGBoost (0.043 vs 0.049 vs 0.069), yang menunjukkan bahwa Random Forest lebih baik dalam hal akurasi relatif untuk setiap prediksi ketika mempertimbangkan skala nilai-nilai tersebut. Pada model Random Forest dan Gradient Boosting memiliki nilai MAPE yang cukup serupa, namun Random Forest sedikit lebih unggul.

**Melihat *goals* dari analisa ini, maka Gradient Boosting akan dipilih** karena lebih unggul besar dalam RMSE, dan juga memiliki nilai MAE dan MAPE yang tidak terlalu jauh dengan Random Forest. Model Gradient Boosting dapat menangani interaksi yang kompleks dari *behaviour* pelanggan dengan lebih baik dan meminimalisir *large deviation* sehingga bisa mencegah *misidentify* pelanggan yang memiliki CLV tinggi. 

> Model Gradient Boosting menjadi pilihan karena memiliki performa lebih baik dari Random Forest dan XGBoost dalam hal akurasi, maka selanjutnya kita akan menggunakan model **Gradient Boosting ini sebagai model akhir**. **Gradient Boosting**

[Diagram](https://www.geeksforgeeks.org/ml-gradient-boosting/) dibawah ini menjelaskan bagaimana *gradient-booted trees* dilatih untuk problem-problem regresi. 
<center><img src="https://media.geeksforgeeks.org/wp-content/uploads/20200721214745/gradientboosting.PNG"  width=800 height=100></center>

reference: [scikit-learn.org : plot gbr](https://scikit-learn.org/1.5/auto_examples/ensemble/plot_gradient_boosting_regression.html), [scikit-learn.org : sklearn.ensemble](https://scikit-learn.org/dev/modules/generated/sklearn.ensemble.GradientBoostingRegressor.html)

Model tree 1 dilatih menggunakan dataset asli (X,y), menghasilkan prediksi tree 1 = y1(hat) dan residual r1 dihitung sebagai r1 = y - y1(hat). Lalu pada model tree 2 dilatih menggunakan dataset residual (X,r1), menghasilkan prediksi tree 2 = r1(hat) dan residual r2 dihitung sebagai r2 = r1 -r1(hat). Proses ini berlanjut hingga jumlah *trees* yang diinginkan tercapai. Setiap tree dilatih untuk memprediksi residual dari tree sebelumnya, sehingga model secara bertahap memperbaiki kesalahan prediksi.

Boosting sendiri adalah teknik ensemble yang menggabungkan hasil dari beberapa model yang lebih sederhana. untuk meningkatkan akurasi prediksi. Model-model ini dilatih secara berurutan, dimana setiap model baru berusaha untuk memperbaiki kesalahan (error) yang dibuat oleh model sebelumnya. Dalam Gradient Boosting, setiap model dilatih untuk mengurangi *loss function* dari model sebelumnya dengan menggunakan pendekatan gradien descent.

Meskipun terkenal dengan akurasinya, kelemahan utama dari model ini adalah waktu pelatihan yang cukup lama dan cukup sensitif terhadap noise.

> Lalu, dengan hasil eva metrics tadi, bagaimana meningkatkan performa model?

**Hyperparameter Tuning**

Pada tahap ini, kita akan melakukan Hyperparameter Tuning pada model Gradient Boosting dengan harapan dapat meningkatkan performa model.

**Hyperparameter pada Gradient Boosting Regression** mengatur proses machine learning gradient boost, yang akan memengaruhi kompleksitas, waktu pelatihan, dan generalisasinya. *Fine-tuning* parameter ini sangat penting untuk kinerja yang optimal. Ada beberapa teknik untuk hyperparameter tuning yaitu **GridSearchCV** dan **RandomizedSearchCV**. Disini kita akan menggunakan **GridSearchCV** untuk Hyperparameter Tuning.

reference : [geeksforgeeks.org : how-to-tune](https://www.geeksforgeeks.org/how-to-tune-hyperparameters-in-gradient-boosting-algorithm/)

```python
# Import library
from sklearn.model_selection import GridSearchCV
```

```python
# Menentukan hyperparam
hyperparam_gbr = {
    'algo__n_estimators': [100, 200, 300, 500], # melakukan trial-error agar hyperparam tidak terlalu banyak
    'algo__learning_rate': [0.01, 0.05, 0.1], 
    'algo__max_depth': [3, 5, 7], 
    'algo__min_samples_split': [2, 5, 10], 
    'algo__min_samples_leaf': [1, 2, 4],
    'algo__subsample': [0.7, 0.8, 0.9, 1.0] 
}

# GridSearchCV
grid_search = GridSearchCV(
    estimator=pipe_model,
    param_grid=hyperparam_gbr,
    cv=5,  # 5-fold cross-validation
    scoring = ['neg_root_mean_squared_error', 'neg_mean_absolute_error', 'neg_mean_absolute_percentage_error'], 
    refit = 'neg_root_mean_squared_error',
    n_jobs=-1,  # Use all processors
    verbose=1
)

# Fit to training data
grid_search.fit(X_train, y_train)

# Display the best parameters and score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("Best Hyperparameters:", best_params)
print("Best CV Score (RMSE):", best_score)
```

**Predict to Test Set with the Tuned Model**

```python
# list evaluation metrics
rmse_gbr_tuned = []
mae_gbr_tuned = []
mape_gbr_tuned = []

# Model XGBoost
model = {'GBR': GradientBoostingRegressor(random_state=0)}

# best method
gbr_tuning = grid_search.best_estimator_

# Fitting model
gbr_tuning.fit(X_train, y_train)

# Predict ke the test set
y_pred_gbr_tuning = gbr_tuning.predict(X_test)

# hitung evaluation metrics
rmse_gbr_tuned_value = np.sqrt(mean_squared_error(y_test, y_pred_gbr_tuning))
mae_gbr_tuned_value = mean_absolute_error(y_test, y_pred_gbr_tuning)
mape_gbr_tuned_value = mean_absolute_percentage_error(y_test, y_pred_gbr_tuning)

# result di append
rmse_gbr_tuned.append(rmse_gbr_tuned_value)
mae_gbr_tuned.append(mae_gbr_tuned_value)
mape_gbr_tuned.append(mape_gbr_tuned_value)

# buat dataframe
tuned_results = pd.DataFrame({
    'Model': 'Gradient Boosting',
    'RMSE Tuned': rmse_gbr_tuned,
    'MAE Tuned': mae_gbr_tuned,
    'MAPE Tuned': mape_gbr_tuned
}).round(3)

# hasil
tuned_results
```
